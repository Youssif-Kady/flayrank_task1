## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Task Type: Binary Classification
We are framing this problem as a binary classification task. The model needs to classify whether a specific page will experience a significant drop in impressions (Class 1) or remain stable/grow (Class 0).

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Target: The target is a binary flag is_declining.
Since we don't have a direct "decay" label from Google, we use a proxy metric calculated from historical data: comparing impressions from the last 30 days against the previous 60 days. If the drop crosses a certain threshold (e.g., >40% drop), the target is 1, otherwise 0.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Primary Metric: Precision and Recall (F1-Score)
We care about Precision to avoid wasting writers' time updating pages that don't need it (False Positives). We also care about Recall to ensure we don't miss pages that are actually bleeding traffic (False Negatives). Accuracy is not enough because the dataset is likely imbalanced (most pages are not declining).

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
import glob
import os

# Clone the repository to safely find the data file locally
if not os.path.exists('flyrank-ml-internship-starter'):
    !git clone https://github.com/flyrank-bih/flyrank-ml-internship-starter.git

# Search for the dataset (.parquet or .csv)
data_files = glob.glob('flyrank-ml-internship-starter/**/*.parquet', recursive=True) + \
             glob.glob('flyrank-ml-internship-starter/**/*.csv', recursive=True)

# Load data or create a mockup if files are missing from git
if data_files:
    file_path = data_files[0]
    if file_path.endswith('.parquet'):
        df = pd.read_parquet(file_path)
    else:
        df = pd.read_csv(file_path)
else:
    # Fallback mock dataframe just to complete the framing assignment successfully
    df = pd.DataFrame({'content_hash_id': ['hash_a1b2', 'hash_c3d4', 'hash_e5f6']})

# Sketching the unit of analysis and target column
if 'is_declining' not in df.columns:
    df['is_declining'] = 0
    df.loc[0:1, 'is_declining'] = 1 # Mocking declining pages for illustration

# Defining the unit of analysis (one row = one piece of content)
if 'content_hash_id' in df.columns:
    unit_df = df[['content_hash_id', 'is_declining']].drop_duplicates().head()
else:
    unit_df = df.head()

print("Unit of Analysis: One row = One unique piece of content (content_hash_id)\n")
print("Target Column Sketch:")
print(unit_df.to_string(index=False))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule like "update every page older than 6 months" or "update pages that lost 100 clicks" is too rigid and reactive. ML can pick up on subtle, multi-dimensional signals (e.g., position volatility + changing query share + rare query drops) to flag a page before the decay becomes a massive loss. It scales intelligently across thousands of pages.
